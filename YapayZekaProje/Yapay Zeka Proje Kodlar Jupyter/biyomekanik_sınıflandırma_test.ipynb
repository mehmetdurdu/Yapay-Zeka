## Plotting Kullanımı
> Aşağıda bulunan örnekte pandas kutüphanesinde bulunan 'Plotting' fonksiyonu nasıl parametre aldığı ve parametrelerin ne iş yaradığı gösterilmektedir. detaylı bilgi için https://pandas.pydata.org/pandas-docs/version/0.15.0/visualization.html bu linkten inceleyebilirsiniz.
- ```dataset["column"].value_counts().plot(autopct="%1.1f%%", kind='pie', figsize=(6, 6))```
- ```dataset["column"]``` : pandas ile okuduğumuz .csv dosyasının içerisindeki sütunu belirtmek, geleneksel python arraylist mantığını kullanırız. örnek olarak ```dataset_2c["class"]``` dataset_2c arraylist in içindeki class sütununa bu şekilde ulaşırız.
- ```.value_counts()``` : seçtiğimiz sütun içerisindeki değerlerin toplamını elde ederiz.
- ```.plot(autopct="%1.1f%%", kind='pie', figsize=(6, 6)``` : plot fonksiyonu ile de oluşturmak istediğimiz grafik türünü, yüzdesi ve görsel boyutu gibi genel parametreler ile verimizi görselleştiriz. ```autopct="%1.1f%%``` bu parametre grafik üzerinde value_counts() fonksiyonundan gelen toplam yüzdeyi verir. ```kind='pie'``` kind parametresi grafik görselinin türünü(pie, bar vs) belirler. ```figsize=(6, 6)``` grafik görselinin boyutunu belirler.

# Model Eğitimi

````python
# Modelin eğtimi ve test işlemi sınıflandırıcılar 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Dataset üzerinde veri önişleme ve bölümleme kütüphaneleri
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
````

````python
# Dataset in X ve Y olarak hedef sütunun seçilmesi
def get_X_and_y(dataset, target_name):
    X=dataset.drop(target_name, axis=1) # dataset te target_name dışında bulunan sütunların X değişkenine atama
    
    y=dataset[target_name] # dataset te sadece target_name in geleceği sütunun y değişkenine atama
    

    X=StandardScaler().fit_transform(X) # X değişkenine atanan sütunların normalizasyon işlemi
    
    labelEncoder=LabelEncoder()     # dataset te bulunan 'class' sütunun string ifade olduğundan 
    y=labelEncoder.fit_transform(y) # dolayı sınıflandırıcının anlayacağı sayısal değere dönüştürme işlemi
    return X, y
````

````python
def predict_model(X, y):
    result_cols = ["Sınıflandırıcılar", "Doğruluk"]  # görsel sınıflandırma sonuçlarının sütun etiketleri
    result_frame = pd.DataFrame(columns=result_cols) 
    
    X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=42) # %70 train %30 test 
    
    # models array içine sınıflandırıların append edilmesi
    # bu sayede bir çok sınıflandırıcıyı tek seferde test edebiliriz
    models = []
    models.append(("NB",GaussianNB()))
    models.append(("KNN",KNeighborsClassifier(n_neighbors = 3)))
    models.append(("SVC", SVC(kernel='linear')))

    for name, model in models:
        model.fit(X_train, y_train) # models array içinde bulunan sınıflandırıcılar ile fit() işlemi
        predicted = model.score(X_test, y_test) #  sınıflandırıcılar ile karşılaştırma 'X_test, y_test'
        
        print (name+' accuracy = '+str(predicted)+'%') # sınıflandırma sonuçlarının ekrana basılması
        acc_field = pd.DataFrame([[name, predicted]], columns=result_cols) # sınfılandırma
        result_frame = result_frame.append(acc_field)                      # sonuçların tabloya basılması

    sns.set_color_codes("muted") # tablodaki bar rengi
    sns.barplot(x='Doğruluk', y='Sınıflandırıcılar', data=result_frame, color="r") # tabloda oranların barplot()
                                                                                   # eklenmesi 
    # tablonun ekrana basılması
    plt.xlabel('Doğruluk %')
    plt.title('Sınıflandırıcı Doğruluğu')
    plt.show()
````

- ## dataset_2c | column_2C.csv

````python
print("dataset_2c")
X, y=get_X_and_y(dataset=dataset_2c, target_name='class')
predict_model(X, y)
````

- ## dataset_3c | column_3C.csv

````python
print("dataset_3c")
X, y=get_X_and_y(dataset=dataset_3c, target_name='class')
predict_model(X, y)
````

# En iyi öznitelik

````python
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import classification_report, confusion_matrix 

def best_attr(dataset, target_name):
    X=dataset.drop(target_name, axis=1) # dataset te target_name dışında bulunan sütunların X değişkenine atama
    y=dataset[target_name] # dataset te sadece target_name in geleceği sütunun y değişkenine atama
    X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=42) # %70 train %30 test 
    best_score=0
    
    # parametre aralığını tanımlama
    gammas=[0.001, 0.01, 0.1, 1, 10, 100]
    Cs=[0.001, 0.01, 0.1, 1, 10, 100]
    kernels=['rbf', 'linear']
    
    # for döngüsündeki asıl amaç best score görebilmek çünkü döngü dışında
    # yapılmış olsayıdı ekrana basılmış olan sonuçlardan bulmak güç olurdu.
    for gamma in gammas:
        for C in Cs: 
            for kernel in kernels:
                svm=SVC(kernel=kernel,gamma=gamma, C=C) # svm sınıflandırıcaya tanımlanan parametre aralığının eklenmesi
                svm.fit(X_train, y_train) # svm sınfılandırıcı ile fit() işlemi
                
                score=svm.score(X_test, y_test) # svm sınıflandırıcı karşılaştırma sonucu(score)

                if score>best_score: # ekrana basılan sonuçlardan en iyi olanı bulma
                    y_pred=svm.predict(X_test) # X_test in predict(tahminleme) işlemi
                    best_score=score # tanımlı olan best_score değişkenine if kararından çıkan sonucu atama
                    best_params={'kernel':kernel, 'C':C, 'gamma':gamma} # en iyi öz niteliği atama
        
    # yukarıdaki yapılan işlemlerin sonunucunda ekrana basılması
    print("best score:",best_score)
    print("best params:",best_params)
    print("classification reports:\n",classification_report(y_test, y_pred))
````

- ## dataset_2c | column_2C.csv

````python
best_attr(dataset=dataset_2c, target_name='class')
````
